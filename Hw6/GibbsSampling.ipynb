```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = 100
b = 99
mean = [0, 0]
cov = [[a, b], [b, a]]

# 相关系数接近1，接近线性关系
data = np.random.multivariate_normal(mean, cov, 50)
x1, x2 = data.T
plt.scatter(x1, x2)
```

```python
# 利用均匀分布在[-20, 20]^2 随机生成50个点，然后利用Gibbs采样迭代
x = np.random.random((50, 2)) * 40 - 20
x1, x2 = x.T
plt.scatter(x1, x2)
```

```python
for _ in range(100):
    x.T[0] = np.random.normal(b/a*x.T[1], a-b**2/a)
    x.T[1] = np.random.normal(b/a*x.T[0], a-b**2/a)

x1, x2 = x.T
plt.scatter(x1, x2)
```
